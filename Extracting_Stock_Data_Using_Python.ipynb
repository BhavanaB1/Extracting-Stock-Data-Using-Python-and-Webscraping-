Importing Tesla data

```python
Tesla=yf.Ticker("TSLA")
```

```python
tesla_data=Tesla.info
```

```python
with open('Tesla.json','w') as openfile1:
  json.dump(tesla_data,openfile1,indent=4)
```

```python
with open('Tesla.json','r') as openfile1:
  json_object1=json.load(openfile1)
print(json_object1)
```

```python
Tesla.info['country']
```

```python
#A share is the single smallest part of a company's stock that you can buy, the prices of these shares fluctuate over time. Using the history() method we can get the share price of the stock over a certain period of time. Using the period parameter we can set how far back from the present to get data. The options for period are 1 day (1d), 5d, 1 month (1mo) , 3mo, 6mo, 1 year (1y), 2y, 5y, 10y, ytd, and max
```

```python
#The format that the data is returned in is a Pandas DataFrame. With the Date as the index the share Open, High, Low, Close, Volume, and Stock Splits are given for each day.
#We can reset the index of the DataFrame with the reset_index function. We also set the inplace paramter to True so the change takes place to the DataFrame itself.
#We can plot the Open price against the Date:
```

```python
#Extracting Share Price
Tesla_share_price_data = Tesla.history(period="max")
Tesla_share_price_data.head()
Tesla_share_price_data.reset_index(inplace=True)
Tesla_share_price_data.plot(x="Date", y="Open")
```

```python
#Extracting Dividends
Tesla.dividends
Tesla.dividends.plot()
```

Importing Gamestop data

```python
Gamestop=yf.Ticker('GME')
```

```python
Gamestop_data=Gamestop.history(period="max")
```

```python
# Convert the DataFrame to a JSON-serializable format using to_json()
# orient='records' creates a list of dictionaries, suitable for JSON
with open ('Gamestop_data.json','w') as openfile3:
  json.dump(Gamestop_data.to_json(orient='records'),openfile3,indent=4)
```

```python
with open('Gamestop_data.json','r') as openfile3:
  json_object3=json.load(openfile3)
print(json_object3)
```
